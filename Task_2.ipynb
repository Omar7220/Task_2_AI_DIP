```python
import pandas as pd
import glob

# List all Excel files in the directory
excel_files = glob.glob(r'D:\AI Diplome\Day 2\archive\*.csv')  

# Initialize an empty list to store DataFrame objects
dfs = []

# Read each Excel file and append its DataFrame to the list
for file in excel_files:
    dfs.append(pd.read_csv(file))

# Concatenate all DataFrames in the list
combined_df = pd.concat(dfs, ignore_index=True)

# Write the combined DataFrame to a new Excel file
combined_df.to_csv(r'D:\AI Diplome\Day 2\archive\combined_file.csv', index=False)  # Change the output file name as needed
```

```python
sample_df_2002 = pd.read_csv (r'D:\AI Diplome\Day 2\archive\FIFA - 2002.csv')
sample_df_2006 = pd.read_csv (r'D:\AI Diplome\Day 2\archive\FIFA - 2006.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Select the columns you want
string_column = 'Team'  
numeric_column = 'Points'

# Extract the selected columns
selected_data = sample_df_2002[[string_column, numeric_column]]

# Plotting
fig, ax1 = plt.subplots(figsize = (10,6) )
plt.xticks(rotation='vertical')


# Plot bars for string values
ax1.scatter(selected_data[string_column], selected_data[numeric_column], color='b', label='Numeric Value')
ax1.set_ylabel('Points ')

# Create a secondary y-axis for the numeric values

plt.xticks(rotation='vertical')

# Show plot
plt.title('points for each team FIFA-2002')
plt.tight_layout()
plt.show()
```

```python
string_column = 'Team'  
numeric_column = 'Goal Difference'

# Extract the selected columns
selected_data = sample_df_2006[[string_column, numeric_column]]

# Plotting
fig, ax1 = plt.subplots(figsize = (10,6) )
plt.xticks(rotation='vertical')

# Plot bars for string values


# Create a secondary y-axis for the numeric values
ax2 = ax1.twinx()
ax2.plot(selected_data[string_column], selected_data[numeric_column], color='r', marker='o', label='String Value')
ax2.set_ylabel('Goal Diffrence ')

plt.xticks(rotation='vertical')

# Show plot
plt.title('Goal Difference for each team FIFA-2006 ')
plt.tight_layout()
plt.show()
```
